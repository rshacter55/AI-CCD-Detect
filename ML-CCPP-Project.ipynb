```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define features and target
all_features = ['AT', 'AP', 'RH', 'V']
target = 'PE'

X_data = df[all_features]
y_target = df[target]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_data, y_target, test_size=0.2, random_state=42)

print("Dataset sizes")
print("-----------------")
print("Training set: ", len(X_train))
print("    Test set: ", len(y_test))

```

# **Req 3: Build and Evaluate LR and RFR Models**



## LINEAR REGRESSION




```python
import math
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
# Crate the model
model_lr = LinearRegression()

# Train the model on the training data
model_lr.fit(X_train, y_train)

# Predict the target values on the test data
y_test_pred_lr  = model_lr.predict(X_test)

# Calculate the mean squared error
mse_test_lr = mean_squared_error(y_test, y_test_pred_lr)

# Calculate the root mean squared error (RMSE)
rmse_lr = mean_squared_error(y_test, y_test_pred_lr, squared=False)

# Calculate the mean absolute error (MAE)
mae_lr = mean_absolute_error(y_test, y_test_pred_lr)

# Calculate MAPE
mape_lr = mean_absolute_percentage_error(y_test, y_test_pred_lr)

# Calculate R-squared (R2) score for Random Forest model
r2_test_lr = r2_score(y_test, y_test_pred_lr)

n = len(y_test)
k = X_data.shape[1]

adjusted_r2_lr = 1 - ((n - 1) / (n - k - 1)) * (1 - r2_test_lr)
print("Adjusted R2",adjusted_r2_lr)

# Calculate the baseline MSE (predicting the mean of target for all instances)
baseline_pred_lr = y_train.mean()  # or use median()
mse_baseline_lr = mean_squared_error(y_test, [baseline_pred_lr] * len(y_test))
# print(f"Baseline Mean Squared Error: {mse_baseline_lr:.2f}")

# Calculate the improvement by comparing baseline MSE with Random Forest MSE
improvement_lr = mse_baseline_lr - mse_test_lr
# print(f"Improvement: {improvement_lr:.2f}")

eval_metrics_lr = [ round(mse_test_lr, 4),  round(r2_test_lr, 4), round(mape_lr, 4), round(rmse_lr, 4),  round(mae_lr, 4), round(improvement_lr, 4)]

print("Linear Regression - Evaluation Metrics")
print("_______________________________________")

# Compare Test and training in table
metrics_df_lr = pd.DataFrame({
     'Names': ['Mean Squared Error (MSE)',  'R-squared (R2)', 'Mean Absolute Percent Error (MAPE)',  'Root Mean Squared Error (RMSE)', 'Mean Absolute Error (MAE)',  'Improvement over Baseline MSE'],
     'Values': eval_metrics_lr
})



print(metrics_df_lr)
```

## **4.2 - hyperparameters**
Given the same algorithm, hyperparameters help find the best configuration of settings for the ML model. Hyperparameters control aspects of the learning process. They are set before training and are not learned from the data. They can have an significant impact on the model's performance.



### **LR - Hyperparameter**
**Ridge Regression - alpha parameter**
Linear Regression does not have hyperparameters to tune. For this testing, I will use a variation on Linear Regression called Ridge Regression. Ridge Regression offers linear regression with regularization. And this hypothetical regularization paramenter ('alpha') is a hyperparameter.

This code seeks the best hyperparameter alpha and outputs the alpha value and the best MSE.

**Fidings**

Optimizing the "alpha" paramenter on the Ridge regression, impoved the performance of the Linear Regression Model only slightly (0.0003 unit) .
- Linear Regression MSE: 20.2737
- Ridge MSE alpha 15  : <font color="green">20.2734</font>

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error


# Split the data into training and testing sets
X_train_rg, X_test_rg, y_train_rg, y_test_rg = train_test_split(X_data, y_target, test_size=0.2, random_state=42)

# Create a Ridge Regression model
model_rg = Ridge()

# Set up the alpha values to test
alphas = [0.01, 0.1, 1.0, 10.0, 15.0]

# Create a parameter grid for GridSearchCV
param_grid = {'alpha': alphas}

# Create GridSearchCV object
grid_search_rg = GridSearchCV(model_rg, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the model using GridSearchCV
grid_search_rg.fit(X_train_rg, y_train_rg)

# Get the best estimator and its corresponding alpha
best_model_rg = grid_search_rg.best_estimator_
best_alpha_rg = best_model_rg.alpha

# Predict the target values on the test data using the best model
y_pred_rg = best_model_rg.predict(X_test_rg)

# Calculate Mean Squared Error
mse_rg = mean_squared_error(y_test, y_pred_rg)
print("Best Alpha:", best_alpha_rg)
print("Best Model MSE:", round(mse_rg, 4))
```

### RFR - hyperparameters

In Random Forest Regression, the hyperparameter n_estimators refers to the number of individual decision trees that are combined to form the ensemble. Each decision tree in the Random Forest is built on a different subset of the training data, and they are aggregated to make predictions.

In Random Forest Regression, the hyperparameter max_depth refers to the maximum depth that each decision tree can grow to during the training process. The depth of a decision tree determines the number of splits or nodes in the tree structure. The number of splits affects the complexity of the tree and its ability to capture complex patterns in the training data.

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor

# Define hyperparameters to tune
param_dist = {
    #'n_estimators': [50, 100, 150],
    'n_estimators': [50, 150],
    'max_depth': [2, 5, 10, 15]
}

# Initialize the Random Forest Regression model
model = RandomForestRegressor(random_state=42)

# Initialize RandomizedSearchCV with 5-fold cross-validation
random_search = RandomizedSearchCV(model, param_distributions=param_dist, n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)

# Perform randomized search
random_search.fit(X_train, y_train)

# Get the best hyperparameters and best estimator
best_params = random_search.best_params_
best_model = random_search.best_estimator_

# Make predictions on the test data using the best model
y_pred = best_model.predict(X_test)

# Calculate MSE
mse = mean_squared_error(y_test, y_pred)
print("Best hyperparameters:", best_params)
print("Best model MSE:", mse)
```

## **Model Evaluation Metrics Summary and Interpretaion**



The evaluation metrics for the Random Forest Regression model suggest that the model is performing reasonably well.
* **Mean Squared Error (MSE)**: The MSE value of **10.52** is relatively low. This indicates that, on average, the squared difference between the predicted and actual values is small. Hence, the model's predictions are close to the actual values.

* **R-squared (R2) Score**: The R-squared value of **0.96** is high. R-squared represents the proportion of the variance in the target variable that is explained by the model. An R2 of 0.96 indicates that the model can explain about 96% of the variance in the target variable, using the given feature t dataset (and model). This is a strong indicator of a well-fitting model.

Other output parameters:

*  **The Mean Absolute Percentage Error (MAPE)**: This is probably the easiest metric to interpret. It measures the average percentage difference between the predicted and actual values. **A measure of 1% (0.01)** means that on average the model is off by 1% from the actual values. It is considered a low MAPE which means that the model's predictions are accurate and have a small relative error compared to the actual values.

* **Improvement mean squared error (MSE) over the baseline MSE** : An "positive" improvement score shows that the model's predictions are better than the baseline. The larger the improvement score, the more value the model is adding in terms of predictive accuracy. The positive value of **269.80** suggests a significant improvement.

Other values:

* **Root Mean Squared Error (RMSE)**: The RMSE value of **4.51** is also relatively low. RMSE gauges the average magnitude of the prediction errors in the original units of the target variable. A lower RMSE is better. A value of 4.51 suggests that the average prediction error is about 4.51 units.

* **Mean Absolute Error (MAE)**: Like RMSE, a lower MAE is better. The MAE value of **2.33** is relatively small, indicating that, on average, the absolute difference between the predicted and actual values is small.



**Conclusion**

Overall, the evaluation metrics suggest that the Random Forest Regression model is performing well. It predicts the target variable pretty accurately. The high R-squared confirms that the model captures a significant portion of the variability in the target variable.
